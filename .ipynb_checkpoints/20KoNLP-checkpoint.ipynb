## KoNLPy
* NLTK는 영어 정보처리를 위한 패키지
* 반면, KoNLPy는 한국어 정보처리를 위한 패키지임
* 서울대학교 산업공학과 개발한 형태소 분석기
     + 이미 개발된 한글 형태소 분석기를 파이썬에서 바로 사용할 수 있도록 도와줌
     + 일종의 래퍼 패키지
* 현재까지 지원하는 형태소분석기는 모두 5가지
     + Hannanum, Kkma, Komoran, mecab, Okt
* konlpy-ko.readthedocs.io / konlpy.org

```python
import konlpy

from konlpy.tag import Okt  # 구 twitter
from konlpy.tag import Kkma
```

```python
# 트위터 형태소 사전을 사용하기 위해 초기화
kkma = Kkma()
twitter = Okt()
```

```python
txt1 = '아버지가 방에 들어가신다.'
txt2 = '나는 보리밥을 먹었다.'
txt3 = '롯데마트가 판매하고 있는 흑마늘 양념 치킨이 논란이 되고 있다.'
```

```python
# 형태소 분석1 : pos, 중요 품사 기반 추출 
twitter.pos(txt1)
```

```python
kkma.pos(txt1)
```

```python
# 형태소 분석2 : morphs, 상세 품사 기반 추출 
twitter.morphs(txt1)
```

```python
kkma.morphs(txt1)
```

```python
# 형태소 분석3 : nouns, 명사 기반 추출 
twitter.nouns(txt1)
```

```python
kkma.nouns(txt1)
```

## 대한민국 헌법 말뭉치 워드클라우드
* 한글폰트는 font_path 속성으로 설정

```python
from nltk import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# fdwords = FreqDist(twitter.nouns(doc1))
fdwords = FreqDist(kkma.nouns(doc1))

wc = dict(fdwords.most_common())
```

```python
wcimg = WordCloud(background_color='white',
    font_path=r'/Library/Fonts/NanumGothic.ttf',
    width=640, height=480).generate_from_frequencies(wc)
plt.imshow(wcimg, interpolation='bilinear')        
plt.axis('off')
plt.show()
```

## 스티브잡스 스탠포드대 졸업축사 워드클라우드

```python
url = 'https://raw.githubusercontent.com/siestageek/datasets/master/txt/stevejobs_ko.txt'

res = requests.get(url)
docs2 = res.text
docs2[:100]
```

```python
# words = Counter(twitter.nouns(docs2))
words = Counter(twitter.nouns(docs2))

wc = dict(words.most_common())
```

```python
wcimg = WordCloud(background_color='white',
    font_path=r'/Library/Fonts/NanumGothic.ttf',
    width=640, height=480).generate_from_frequencies(wc)
plt.imshow(wcimg, interpolation='bilinear')        
plt.axis('off')
plt.show()
```

# 색상팔레트 유형
    - sequential : 수치데이터, 순서가 있는 데이터
    - qualitative : 범주형 데이터
    - diverging : 수치형/범주형 데이터

```python
pip install palettable
```

```python
pip install colorbrewer
```

```python
import random
from palettable.colorbrewer.sequential import Reds_9
from palettable.colorbrewer.diverging import RdYlBu_11
from palettable.colorbrewer.qualitative import Pastel2_8
```

## 워드클라우드에 사용할 함수 정의

```python
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    # return tuple(RdYlBu_11.colors[random.randint(0, 10)])
    # return tuple(Reds_9.colors[random.randint(0, 8)])
    return tuple(Pastel2_8.colors[random.randint(0, 7)])

```

```python
wcimg = WordCloud(background_color='white',
    font_path=r'/Library/Fonts/NanumGothic.ttf',
    width=640, height=480)\
    .generate_from_frequencies(wc)\
    .recolor(color_func=color_func) # 색상 칠하기

plt.imshow(wcimg, interpolation='bilinear')        
plt.axis('off')
plt.show()
```
