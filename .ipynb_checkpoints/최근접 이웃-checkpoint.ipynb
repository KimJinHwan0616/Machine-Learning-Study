## KNN 
* k 최근접 이웃(k-nearest neighbors) 알고리즘
* 머신러닝 분류에 자주 사용되는 대표 알고리즘
* 얼굴인식, 개인영화추천, 질병 유전자 패턴 식별에 활용
* KNN의 K는 가장 가까운 이웃 '하나'가 아니고 훈련 데이터 중 새로운 데이터와 가장 가까운 k개의 이웃을 찾는다는 의미
* 즉, 하나의 관측값은 거리가 가까운 k개의 이웃 관측값들과 비슷한 특성을 갖는다고 가정함
    + 거리를 구할때는 유클리드 거리, 맨해튼 거리, 코사인 유사도, 피어슨 상관계수등이 사용됨
* 따라서, K개 이웃의 목표변수 중 다수결로 가장 많은 범주에 속한 값을 결과로 반환
* KNN 알고리즘에서는 k를 얼마로 설정하느냐에 따라 결과와 성능이 달라짐
    + k가 작으면 데이터의 범위가 좁아짐 - 과적합 위험
    + k가 크면 데이터의 범위가 넓어짐 - 일반화 위험
    + 일반적으로 k값은 데이터 건수에 제곱근을 씌운 값

# 과일, 채소 구분하기
* KNN 알고리즘을 이용해서 당도, 아삭함을 기준으로 과일인지 채소인지 구분
    + 당도가 6, 아삭함이 4 인 토마토는 채소인가 과일인가?


```python
fresh = pd.read_csv('../data/fresh.csv', encoding='euc-kr')
fresh.columns = ['name','sweet','crunchy','class']

fresh.head()
```

```python
plt.plot(fresh.sweet, fresh.crunchy, 'ro')
plt.show()
```

```python
sns.scatterplot(data=fresh, x='sweet', y='crunchy', hue='class')
plt.plot(6,4,'ro')
for i, n in enumerate(fresh.name):
    plt.annotate(n, (fresh.sweet[i], fresh.crunchy[i]))

plt.annotate('토마토는 무엇으로 분류될까', (6,4))
plt.grid()
plt.show(True)
```

```python
# 데이터 전처리 : map
fresh['target'] = fresh['class'].map(
    {'과일':0, '단백질':1, '채소':2})
fresh.head()
```

```python
# 데이터 분할
data = fresh.iloc[:, 1:3]
target = fresh.target
```

```python
# 모델 학습
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(data, target)
knn.score(data, target)   
```

```python
# 예측
one = [[6, 4]]
knn.predict(one)
```

# 적절한 k값 찾기
* n_neighbors 값을 설정했을때의 정확도 조사

```python
scores = []
for k in range(1, 10+1):
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, data, target, cv=5, scoring='accuracy')
    scores.append(np.mean(score))
    
plt.plot(range(1, 10+1), scores, 'ro--')
plt.show()

# 과적합 제외, 
```

# 포지션을 결정짓는 요인 파악
- 센터/슈팅가드를 구분지을 명확한 경계 없음: 2점슛, 스틸, 어시스트
- 센터/슈팅가드를 구분지을 명확한 경계 있음: 3점슛, 리바운드, 블로킹
    - 단, 스틸, 블로으로 조합했을때도 경계 명확한 경계가 존재
    - 하지만, 슈팅가드/센터라는 포지션을 구분하기에 적당한 속성이라 보기 힘듬

```python
# 시각화1 ( 스틸/2점슛 )
sns.scatterplot(data=bbp, x='STL', y='2P', hue='Pos')
```

```python
# 시각화2 ( 어시스트/2점슛 )
sns.scatterplot(data=bbp, x='AST', y='2P', hue='Pos')
```

```python
# 시각화3 ( 블로킹/3점슛 )
sns.scatterplot(data=bbp, x='BLK', y='3P', hue='Pos')
```

```python
# 시각화4 ( 리바운드/2점슛 )
sns.scatterplot(data=bbp, x='TRB', y='3P', hue='Pos')
```

```python
#  feature 간에 관계를 scatter로 시각화
k = 0
cols = bbp.columns[2:8]
n = len(cols)
if int(((n-1) * n / 2)  % 3):
    row = int((n-1) * n / 2 / 3) + 1
else:
    row = int((n-1) * n / 2 / 3)

plt.figure(figsize=(16, 25))

for i, col in enumerate(cols):
    for col2 in cols[i+1:]:
        k += 1
        plt.subplot(row,3,k) 
        sns.scatterplot(data=bbp, x=col, y=col2, hue='Pos')
plt.show()
```

```python
# 데이터 나누기
data = bbp.loc[:, ['3P', 'BLK']]
bbp['target'] = bbp['Pos'].map({'C':0, 'SG':1})

target = bbp.target
```

```python
Xtrain, Xtest, ytrain, ytest = \
    train_test_split(data, target, test_size=0.3, stratify=target, random_state=2208301305)
```

```python
# 모델 훈련 및 평가
knn = KNeighborsClassifier(n_neighbors=5)

ss = StandardScaler()

Xtrain_scaled = ss.fit_transform(Xtrain)    # 훈련 데이터 → 표준화
Xtest_scaled = ss.transform(Xtest)          # 테스트 데이터 → 표준화

knn.fit(Xtrain_scaled, ytrain)

pred = knn.predict(Xtest)
print(knn.score(Xtrain, ytrain), accuracy_score(ytest, pred))
```

```python
# 모델 훈련 및 평가
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(Xtrain, ytrain)

pred = knn.predict(Xtest)
print(knn.score(Xtrain, ytrain), accuracy_score(ytest, pred))
```

```python
# 예측
p = [[2.5, 2], [5, 0]]
knn.predict(p)
```

```python
# 모델 훈련 및 평가 2 - 교차검증, 적절한 k는?
scores = []
for k in range(1, 10+1):
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, Xtrain, ytrain, cv=5, scoring='accuracy')
    
    scores.append(np.mean(score))
    
plt.plot(range(1, 10+1), scores, 'ro--')
plt.show()

# 과적합 제외, 
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(Xtrain, ytrain)

pred = knn.predict(Xtest)
print(knn.score(Xtrain, ytrain), accuracy_score(ytest, pred))
```
